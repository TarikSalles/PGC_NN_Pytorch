```python
def fazer_histograma(df1,df2,descricao1,descricao2):
  plt.hist(df1, alpha=0.5, label=descricao1)
  plt.hist(df2, alpha=0.5, label=descricao2)
  plt.xlabel('Valores')
  plt.ylabel('Frequência')
  plt.title('Histograma dos Dados')
  plt.legend(loc='upper right')
  plt.show()

def comparar_graficamente(dict1, dict2, descricao1, descricao2, tipo):
    categorias = list(dict1.keys())
    valores_dict1 = list(dict1.values())
    valores_dict2 = list(dict2.values())

    posicoes = np.arange(len(categorias))

    largura_barra = 0.35

    plt.figure(figsize=(10, 6))
    plt.bar(posicoes - largura_barra/2, valores_dict1, width=largura_barra, label=descricao1, color='blue')
    plt.bar(posicoes + largura_barra/2, valores_dict2, width=largura_barra, label=descricao2, color='orange')

    plt.xlabel('Categorias')
    plt.ylabel('Valores')
    plt.title('Comparação métrica média ' + tipo + ':')
    plt.xticks(posicoes, categorias)
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
def calcular_medias_por_coluna(nome_arquivo):
    df = pd.read_csv(nome_arquivo)
    df = df.astype(float)

    medias = df.mean().tolist()

    rotulos_colunas = df.columns.tolist()

    medias_com_rotulos = dict(zip(rotulos_colunas, medias))

    return medias_com_rotulos


```

<h1> Precisão </h1>

```python
precisao_pytorch_media = calcular_medias_por_coluna("Metricas_CSV/pytorch/precision.csv")
precisao_tensorflow_media = calcular_medias_por_coluna("Metricas_CSV/tensorflow/precision.csv")
comparar_graficamente(precisao_pytorch_media, precisao_tensorflow_media,"pytorch","tensorflow","precision")
```

<h1> Recall </h1>

```python
precisao_pytorch_media = calcular_medias_por_coluna("Metricas_CSV/pytorch/recall.csv")
precisao_tensorflow_media = calcular_medias_por_coluna("Metricas_CSV/tensorflow/recall.csv")
comparar_graficamente(precisao_pytorch_media, precisao_tensorflow_media,"pytorch","tensorflow","recall")
```
